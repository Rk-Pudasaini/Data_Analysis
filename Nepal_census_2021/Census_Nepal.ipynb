```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import data_table
```

```python

df = pd.read_csv('/content/Census_Nepal_2021.csv')
df.head()
```

```python
#get the rows and column of the data
df.shape
```

```python
#get the columns names
column_names = df.columns
print(column_names)
```

```python
#rename the column as desired
new_column_names = ['Id','District','Local_Level_Name','Total_Family_No','Total_Household', 'Total_Population', 'Total_Male', 'Total_Female']
df.columns = new_column_names
column_names = df.columns
print(column_names)
```

```python
df.head()
```

```python
#get the missing values of the dataframe
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#find Totals
total_families = df['Total_Family_No'].sum() # number of families specifically counts familial units
print("Total number of families in Nepal till 2021 Census:", total_families)

total_households = df['Total_Household'].sum()
print("Total number of Households in Nepal till 2021 Census:", total_households) #count of households
#A household is defined as a group of people who live together in a housing unit or dwelling and share common living arrangements.
#A household can consist of a single person living alone, a family unit, or a group of unrelated individuals sharing living quarters.
```

```python
#Get the total populations of the country till 2078
total_population  = df['Total_Population'].sum()
print("Total populations of Nepal till 2021 census", total_population)
```

```python
#piechart for total males and females
# Data
total_male = df['Total_Male'].sum()
total_female = df['Total_Female'].sum()

# Pie chart
labels = ['Male', 'Female']
sizes = [total_male, total_female]
colors = ['blue', 'red']
#explode = (0.1, 0)  # explode the 1st slice (Male)

# Plot
plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Male and Female % of total populations')
plt.show()
```

```python
# Filter the DataFrame for Kathmandu district
kathmandu_rows = df[df['District'] == 'Kathmandu']

# Print the filtered rows
print(kathmandu_rows)
```

```python
# Filter the DataFrame for Kageshwori Manahora Municipality
kageshwori_manahora_info = df[df['Local_Level_Name'] == 'Kageshwori Manahora Municipality']

# Print the filtered information
print(kageshwori_manahora_info)
```

```python
#bargraph of total male and female of kageshwori monahara
# Get the total male and female population
total_male = kageshwori_manahora_info['Total_Male'].sum()
total_female = kageshwori_manahora_info['Total_Female'].sum()

# Create bar graph
labels = ['Male', 'Female']
values = [total_male, total_female]
colors = ['lightblue', 'lightcoral']

plt.bar(labels, values, color=colors)
plt.xlabel('Gender')
plt.ylabel('Population')
plt.title('Total Male and Female Population in Kageshwori Manahara Municipality')
plt.show()
```

```python
# Filter the DataFrame for District Kathmandu
kathmandu_population = df[df['District'] == 'Kathmandu']['Total_Population'].sum()

print("Total population of district Kathmandu:", kathmandu_population)
```

```python
# Calculate total population by district
district_population = df.groupby('District')['Total_Population'].sum()

# Sort districts by total population in descending order
sorted_districts = district_population.sort_values(ascending=False)

# Select the top ten districts
top_ten_districts = sorted_districts.head(10)

print("Top ten populations by district:")
print(top_ten_districts)
```

```python
# Plot the bar graph
plt.figure(figsize=(4, 4))
top_ten_districts.plot(kind='bar', color='skyblue')
plt.title('Top Ten Populations by District')
plt.xlabel('District')
plt.ylabel('Total Population in 100K')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlap of labels
plt.show()
```

```python
# Select the top ten  low population districts
top_ten_districts = sorted_districts.tail(10)

print("Top ten low populations by district:")
print(top_ten_districts)
```

```python
# Plot the bar graph
plt.figure(figsize=(4, 4))
top_ten_districts.plot(kind='bar', color='skyblue')
plt.title('Top Ten Low Populations by District')
plt.xlabel('District')
plt.ylabel('Total Population')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlap of labels
plt.show()
```

```python
# Group by 'District' and calculate total population
population_by_district = df.groupby('District')['Total_Population'].sum().sort_values(ascending=False)

# Plot the population distribution
plt.figure(figsize=(12, 8))
population_by_district.plot(kind='bar', color='skyblue')
plt.title('Population Distribution by District')
plt.xlabel('District')
plt.ylabel('Total Population')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlap of labels
plt.show()
```

```python
# Group by 'District' and calculate total population
population_by_district = df.groupby('District')['Total_Population'].sum().sort_values(ascending=False)

# Plot the population distribution with logarithmic scale on y-axis
plt.figure(figsize=(12, 8))
population_by_district.plot(kind='bar', color='skyblue')
plt.title('Population Distribution by District')
plt.xlabel('District')
plt.ylabel('Total Population (log scale)')
plt.yscale('log')  # Set y-axis scale to logarithmic
plt.yticks([10**i for i in range(int(np.log10(population_by_district.min())), int(np.log10(population_by_district.max()))+1)],
           [f'{10**i:,}' for i in range(int(np.log10(population_by_district.min())), int(np.log10(population_by_district.max()))+1)])
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlap of labels
plt.show()
```
